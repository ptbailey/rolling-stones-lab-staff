## Import both datasets (top 500 albums AND top 500 songs)

```python
# dataset : top 500 albums
import pandas as pd
recs = pd.read_csv('./data.csv')
albums = recs.to_dict('recs')
albums # with format keys = number,year,album,artist,genre,subgenre
```

```python
# dataset : top 500 songs
text_file = open('top-500-songs.txt', 'r')
lines = text_file.readlines()

# clean up file and make the list of lists into dictionaries
split_lines = list(map(lambda string: string.rstrip('\n'),lines))
actual_lines = list(map(lambda sline : sline.split('\t'),split_lines))
actual_lines

keys = ('rank', 'name', 'artist', 'year')
top_500_songs = []

for index in actual_lines:
    values = tuple(index)
    dictionary = dict(zip(keys,values))
    top_500_songs.append(dictionary)

for track in top_500_songs:
    track['rank'] = int(track['rank'])
    track['year'] = int(track['year'])
top_500_songs;
```

## Find by name - Takes in a string that represents the name of an album, and list of music. Should return a dictionary with the correct album, or return None.


```python
## format of albums: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def find_by_name(string, music_list):
    for record in music_list:
        try: 
            if (string == record['album']):
                return record
        except KeyError:
            if (string == record['name']):
                return record
```

```python
print(find_by_name("What's Going On", albums))
print(find_by_name("Purple Haze",top_500_songs))
```

## Find by rank - Takes in a number that represents the rank in the list of top albums, and music list,  and returns the album with that rank. If there is no album with that rank, it returns None.


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def find_by_rank(rank, music_list):
    for record in music_list:
        try:
            if rank == record['number']:
                return record
        except KeyError:
            if rank == record['rank']:
                return record
```

```python
print(find_by_rank(6,albums))
print(find_by_rank(6,top_500_songs))
```

## Find by year - Takes in a number for the year in which an album was released and returns a list of albums that were released in that year. If there are no albums released in the given year, it returns an empty list.


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def find_by_year(year, music_list):
    in_year = []
    for record in music_list:
        if year == record['year']:
            in_year.append(record)
            continue
    return in_year
```

```python
print(len(find_by_year(1965,albums)))
print(len(find_by_year(1970,top_500_songs)))
```

## Find by years - Takes in a start year and end year. Returns a list of all albums that were released on or between the start and end years. If no albums are found for those years, then an empty list is returned.


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def year_range(first_year, last_year, music_list):
    rangE = list(range(first_year, last_year+1))
    vinyls = []
    for record in music_list:
        if record['year'] in rangE:
            vinyls.append(record)
            continue
    return vinyls
```

```python
print(len(year_range(1967,1967,albums)))
print(len(year_range(1967,1967,top_500_songs)))
```

## Find by ranks - Takes in a start rank and end rank. Returns a list of albums that are ranked between the start and end ranks. If no albums are found for those ranks, then an empty list is returned.


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def rank_range(first_rank, last_rank, music_list):
    rang = list(range(first_rank, last_rank+1))
    ranks = []
    for record in music_list:
        try:
            if record['number'] in rang:
                ranks.append(record)
                continue
            return ranks
        except KeyError:
            if record['rank'] in rang:
                ranks.append(record)
                continue
            return ranks
```

```python
print(rank_range(1,10, albums));
print(rank_range(1,10,top_500_songs));
```

## All titles - Returns a list of titles for each album.


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def all_titles(music_list):
    names = []
    for record in music_list:
        try:
            names.append(record['album'])
        except KeyError:
            names.append(record['name'])
    return names
    
```

```python
print(all_titles(albums)[0:3])
print(all_titles(top_500_songs)[0:3])
```

## All artists - Returns a list of artist names for each album.

```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def all_artists(music_list):
    artists = []
    for record in music_list:
        artists.append(record['artist'])
    return artists
```

## Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def most_albums(music_list):
    group_artists = all_artists(music_list)
    unique_set_artists = set(group_artists)
    unique_artists = dict.fromkeys(unique_set_artists, 0) # makes dictionary of unique artist
    
    for artist in group_artists:
        unique_artists[artist] += 1
    
    sorted_list_artists = sorted(unique_artists, key = lambda i: unique_artists[i],reverse=True)
    values = sorted(unique_artists.values(),reverse = True)
    max_values = max(values)
    ordered_list = list(zip(sorted_list_artists, values))
    
    most_album_artist = []
    for part in ordered_list:
        if part[-1] == max_values:
            most_album_artist.append(part)
    return most_album_artist
    
```

```python
print(most_albums(albums))
print(most_albums(top_500_songs))
```

## Most popular word - Returns the word used most in amongst all album titles unique_titles = set(titles) # take the list of titles and only show titles once


```python
## format of records: number, year, album, artist, genre, subgenre
## format of top 500 songs: 'rank', 'name', 'artist', 'year'
    # number = rank, name = album, 
```

```python
def popular_word(music_list):
    try:
        list_of_titles = list(map(lambda record: record['album'], music_list))
    except KeyError:
        list_of_titles = list(map(lambda record: record['name'], music_list))
    
    list_of_words = []   
    
    for title in list_of_titles:
        split_title = title.split(' ')
        list_of_words += split_title
    
    unique_words = set(list_of_words)
    words_and_count = dict.fromkeys(unique_words,0) #makes dictionary of unique artist --> set
    
    for word in list_of_words:
        words_and_count[word] += 1
    
    maximum_word_use = max(words_and_count, key = words_and_count.get)
    
    return maximum_word_use, words_and_count[maximum_word_use]
```

```python
print(popular_word(albums))
print(popular_word(top_500_songs))
```
